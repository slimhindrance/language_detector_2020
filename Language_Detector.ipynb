```python
from __future__ import unicode_literals
import pandas as pd
import bs4
import requests
import youtube_dl
#import csv
import json
```

### Read CSV into DataFrame and create list 'languages'

```python
langfile = pd.read_csv('languages.csv')
languages = list(langfile['ISO language name'])
languages
```

## Pull Audio Files from YouTube

```python

#pulls audio files
base = "https://www.youtube.com/results?search_query="

subtit = "+movies+with+english+subtitles"



for files in range (0, 100):
    for langnum in range(0,1):
        r = requests.get(base + languages[langnum] + subtit)

        page = r.text
        soup = bs4.BeautifulSoup(page, 'html.parser')

        vids = soup.findAll('a', attrs={'class': 'yt-uix-tile-link'})

        video_list = []
        for v in vids:
            tmp = 'https://www.youtube.com' + v['href']
            video_list.append(tmp)

        ydl_opts = {
            'format': 'bestaudio/best',
            'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'wav',
            'preferredquality': '192'
        }],
            'postprocessor_args': [
            '-ar', '16000'
        ],
            'prefer_ffmpeg': True,
            'keepvideo': False
        }

        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.download(video_list)
        print(languages[langnum]+' complete')
```
